```python
import pandas as pd
from fbprophet import Prophet
from matplotlib import pyplot as plt 
```

```python
# load 
for i in ["TSLA","MSFT","ATVI","AMZN","GOOGL","ZM","NFLX"]:
    #rename variables
    str_ = i + ".pkl"
    x = pd.read_pickle(str_)
    x = x.reset_index()
    x["ds"] = x["variable"]
    x["y"] = x["4. close"]
    x["rsi"] = x["value"]
    x = x[["ds","y"]]
    m = Prophet(interval_width=.90)
    m.fit(x)
    future = m.make_future_dataframe(periods=10)
    forecast = m.predict(future)
    fig = m.plot(forecast)
    # set plot parameters
    ax = fig.gca()
    ax.set_title(i,size=12)
    ax.set_xlabel("Date",size=10)
    ax.set_ylabel("Stock Price",size=10)
    ax.grid(axis='y',which='minor')
    plt.show()
    # print predicted values
    today = x.head(1)["y"]
    print("Today's actual:",today)
    print(forecast.tail(11)[["ds","yhat_lower","trend","yhat_upper"]])



```
